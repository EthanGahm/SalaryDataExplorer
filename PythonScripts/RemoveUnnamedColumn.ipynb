```python
import pandas as pd
def RemoveUnnamedColumn(filename):
    df = pd.read_csv('../Data/' + filename)
    df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
    df.to_csv("../Data/FinalData2021.csv", index=False)
RemoveUnnamedColumn('FinalData2021_Fixed_Weird_Rows.csv')
```
