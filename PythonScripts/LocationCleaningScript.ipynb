```python
## this method checks if a certain string contains no alphanumeric characters.
def no_alphanumeric(str):
    l = len(str)
    count =0
    for letter in str:
        if not letter.isalnum():
            count = count +1
    if l == count:
        return True
    else:
        return False
## this method checks if a certain string contains only number characters.
def only_numbers(str):
    l = len(str)
    count =0
    for letter in str:
        if  letter.isdecimal():
            count = count +1
    if l == count:
        return True
    else:
        return False
```
