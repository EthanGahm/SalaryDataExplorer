```python
import pandas
from ipynb.fs.full.RemoveSpacesFromHeaders import *
```

```python
# In the "Work Experience" column, the "5 - 7 years" option
# is listed without spaces as "5-7 years"
# This function simply replaces all instances of "5-7 years"
# with "5 - 7 years"
def Normalize5to7(filename):
    df = pandas.read_csv('./' + filename+'.csv')
    df.replace(to_replace = ['5-7 years'], value = '5 - 7 years')
    return df
```

```python
# Also in the "Work Experience" column, values < 1 year are
# written as "1 year or less." 
# This function replaces all instances of "1 year or less"
# with "0 - 1 years"
def Normalize1YearOrLess(df):
    df.replace(to_replace = ['1 year or less'], value = '0 - 1 years')
    return df
```

```python
def NormalizeExperience(filename):
    df = pandas.read_csv(filename)
    df.replace(to_replace = ['5-7 years'], value = '5 - 7 years')
    df.replace(to_replace = ['1 year or less'], value = '0 - 1 years')
    return df
    

NormalizeExperience("../2021_Results.csv")
```
